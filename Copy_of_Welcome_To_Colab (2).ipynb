10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

-> reduce():reduce function takes 2 arguments and reduce a sequence to a single value.

```
'''Map'''
l1=[2,4,5,6,7,8]
print(list(map(lambda x:x**2,l1)))
'''Reduce'''
from functools import reduce
print(reduce(lambda x,y:x+y,l1))

'''Filter'''
list(filter(lambda x:x%2==0,l1))
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```
def final_sum(list1):
  return sum(list1)
def func(n):
  list1=[]
  for i in n:
    if i%2==0:
      list1.append(i)
  return final_sum(list1)
func([1,2,4,5,7,6,9,8])
```

each number.

```
def func(n):
  l1=[]
  for i in n:
    if type(i)==int:
      i=i**2
      l1.append(i)
  return l1

func([2,4,5,6,7,"Naman"])
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```
def fun(n):
  if 1<=n<=200:
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

  else:
    return "out of range"

fun(201)



```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```
def func(n):
  if n==0:
    return 1
  else:
    for i in range(1,n+1):
      yield 2**i

f=func(5)
next(f)
next(f)





```
